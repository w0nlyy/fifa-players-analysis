Opis algorytmów
1. PCA (Principal Component Analysis)
Opis: PCA to liniowa metoda redukcji wymiarowości, która przekształca dane do nowego układu współrzędnych, w którym pierwsza składowa główna wyjaśnia największą wariancję w danych, druga składowa – kolejną największą, i tak dalej. Metoda bazuje na obliczeniu wartości i wektorów własnych macierzy kowariancji lub korelacji danych.

Zalety:

Zmniejsza wymiarowość danych, zachowując jak najwięcej informacji (wariancji).
Szybka i łatwa w implementacji.
Umożliwia wizualizację wielowymiarowych danych w 2D lub 3D.
Wady:

Zakłada liniowe relacje między zmiennymi.
Wynikowe składowe główne są trudne do bezpośredniej interpretacji, ponieważ są liniowymi kombinacjami zmiennych.
Wymaga standaryzacji danych przy zmiennych o różnych skalach.
2. Klasyczne MDS (Classical Multidimensional Scaling)
Opis: Klasyczne skalowanie wielowymiarowe (MDS) odwzorowuje rzeczywiste odległości między punktami w przestrzeni niskowymiarowej, minimalizując zniekształcenia rzeczywistych odległości w danych wejściowych. Metoda opiera się na dekompozycji wartości własnych macierzy odległości.

Zalety:

Zachowuje rzeczywiste odległości między punktami, co pozwala na wizualizację danych w ich przestrzennych relacjach.
Intuicyjnie przedstawia dane w formie mapy odległości.
Wady:

Ograniczona do metrycznych danych wejściowych (np. odległości euklidesowe).
Może być obliczeniowo kosztowna dla dużych zbiorów danych.
Wyniki mogą być wrażliwe na szum w danych.
3. Niemetryczne MDS (Non-Metric Multidimensional Scaling)
Opis: Niemetryczne MDS skupia się na zachowaniu porządku rangowego odległości między punktami zamiast rzeczywistych wartości odległości. Metoda iteracyjnie minimalizuje błąd odwzorowania (stress).

Zalety:

Dobrze nadaje się do danych nieliniowych i zmiennych na skali porządkowej.
Może być bardziej elastyczne niż klasyczne MDS w przypadku danych nienumerycznych.
Wady:

Iteracyjny charakter metody powoduje większe koszty obliczeniowe w porównaniu do klasycznego MDS.
Wynikowe mapy mogą być trudniejsze do interpretacji, gdy porządek rangowy ma niewielki związek z rzeczywistymi odległościami.

```python
# Inicjalizacja PCA
pca = PCA()

# Przekształcenie danych za pomocą PCA
pca_transformed = pca.fit_transform(data_cleaned)

# Wyjaśniona wariancja przez każdą składową główną
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance_ratio)

# Wyświetlenie wariancji wyjaśnionej przez poszczególne składowe
print("Wyjaśniona wariancja przez każdą składową główną:")
for i, var in enumerate(explained_variance_ratio):
    print(f"PC{i+1}: {var:.2%}")

print("\nSkumulowana wariancja wyjaśniona przez kolejne składowe:")
for i, cum_var in enumerate(cumulative_variance):
    print(f"PC{i+1}: {cum_var:.2%}")

# Wizualizacja skumulowanej wariancji
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o')
plt.title("Skumulowana proporcja wyjaśnionej wariancji przez PCA")
plt.xlabel("Liczba składowych głównych")
plt.ylabel("Skumulowana wariancja")
plt.grid()
plt.show()
```

```python
# Obliczenie ładunków czynnikowych
loadings = pca.components_.T

# Tworzenie DataFrame z ładunkami czynnikowymi
loadings_df = pd.DataFrame(loadings, columns=[f'PC{i+1}' for i in range(loadings.shape[1])], index=data_cleaned.columns)

# Wyświetlenie ładunków czynnikowych w formie tabeli
print("Ładunki czynnikowe (wektory własne):")
display(loadings_df)
```

```python
# Obliczenie ładunków czynnikowych
loadings = pca.components_.T

# Tworzenie wykresu
plt.figure(figsize=(8, 8))
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)
for i, var in enumerate(data_cleaned.columns):
    plt.arrow(0, 0, loadings[i, 0], loadings[i, 1], color='b', alpha=0.7)
    plt.text(loadings[i, 0] * 1.1, loadings[i, 1] * 1.1, var, color='r', ha='center', va='center')

plt.xlabel(f"PC1 ({pca.explained_variance_ratio_[0]:.2%})")
plt.ylabel(f"PC2 ({pca.explained_variance_ratio_[1]:.2%})")
plt.title("Wykres ładowań czynnikowych")
plt.grid()
plt.show()

# Ładunki czynnikowe w tabeli
loadings_df = pd.DataFrame(loadings[:, :2], columns=['PC1', 'PC2'], index=data_cleaned.columns)
print(loadings_df)
```

```python


# Obliczenie ładunków czynnikowych
loadings = pca.components_.T

# Tworzenie interaktywnego wykresu 3D
fig = go.Figure()

# Dodanie strzałek dla każdej zmiennej
for i, var in enumerate(data_cleaned.columns):
    fig.add_trace(
        go.Scatter3d(
            x=[0, loadings[i, 0]],
            y=[0, loadings[i, 1]],
            z=[0, loadings[i, 2]],
            mode='lines+text',
            line=dict(color='blue', width=4),
            text=[None, var],
            textposition="top center",
            showlegend=False
        )
    )

# Ustawienie osi i tytułu
fig.update_layout(
    scene=dict(
        xaxis_title=f"PC1 ({pca.explained_variance_ratio_[0]:.2%})",
        yaxis_title=f"PC2 ({pca.explained_variance_ratio_[1]:.2%})",
        zaxis_title=f"PC3 ({pca.explained_variance_ratio_[2]:.2%})",
    ),
    title="Interaktywny Wykres Ładowań Czynnikowych - PCA 3D",
)

fig.show()
```

```python
# Obliczanie loadings (wkłady zmiennych w składowe główne)
loadings = pca.components_.T * np.sqrt(pca.explained_variance_)

# Biplot
plt.figure(figsize=(10, 8))
plt.scatter(pca_transformed[:, 0], pca_transformed[:, 1], alpha=0.5, label="Obserwacje")

# Rysowanie wektorów (loadings)
scale_factor = 3  # współczynnik skalujący
for i, var in enumerate(data_cleaned.columns):
    plt.arrow(
        0, 0, 
        loadings[i, 0] * scale_factor, loadings[i, 1] * scale_factor, 
        color="r", alpha=0.5, head_width=0.5, linewidth=2
    )
    plt.text(
        loadings[i, 0] * scale_factor * 1.15, 
        loadings[i, 1] * scale_factor * 1.15, 
        var, color="black", fontsize=12
    )

plt.title("Biplot: PCA (PC1 vs PC2)")
plt.xlabel("Pierwsza składowa główna (PC1)")
plt.ylabel("Druga składowa główna (PC2)")
plt.axhline(0, color="grey", linestyle="--", linewidth=0.7)
plt.axvline(0, color="grey", linestyle="--", linewidth=0.7)
plt.legend()
plt.grid()
plt.show()
```

```python
# Wybieranie 20% danych
data_sample = data_cleaned.sample(frac=1/16, random_state=42)

# Klasyczne MDS
mds = MDS(n_components=2, random_state=42, metric=True)
mds_transformed = mds.fit_transform(data_sample)

# Wizualizacja klasycznego MDS
plt.figure(figsize=(8, 6))
plt.scatter(mds_transformed[:, 0], mds_transformed[:, 1], alpha=0.6)
plt.title("Klasyczne skalowanie wielowymiarowe (MDS) na 20% danych")
plt.xlabel("Wymiar 1 (Dim1)")
plt.ylabel("Wymiar 2 (Dim2)")
plt.grid(True)
plt.show()
```

```python
# Klasyczne MDS z 3 wymiarami
mds = MDS(n_components=3, random_state=42, metric=True)
mds_transformed = mds.fit_transform(data_sample)

# Tworzenie interaktywnego wykresu 3D
fig = go.Figure()

# Dodanie punktów do wykresu
fig.add_trace(
    go.Scatter3d(
        x=mds_transformed[:, 0],  # Wymiar 1
        y=mds_transformed[:, 1],  # Wymiar 2
        z=mds_transformed[:, 2],  # Wymiar 3
        mode='markers',
        marker=dict(
            size=3,
            color=np.arange(len(mds_transformed)),  # Kolory w zależności od indeksu
            colorscale='Viridis',  # Skala kolorów
            opacity=0.5
        ),
        text=[f"Punkt {i+1}" for i in range(len(mds_transformed))],  # Etykiety punktów
        name='Punkty'
    )
)

# Ustawienia osi
fig.update_layout(
    scene=dict(
        xaxis_title="Wymiar 1 (Dim1)",
        yaxis_title="Wymiar 2 (Dim2)",
        zaxis_title="Wymiar 3 (Dim3)",
    ),
    title="Interaktywny Wykres 3D - Klasyczne MDS",
    margin=dict(l=0, r=0, b=0, t=40),
)

# Wyświetlenie wykresu
fig.show()
```

```python
# Niemetryczne MDS
mds_nonmetric = MDS(n_components=2, random_state=42, metric=False)
mds_nonmetric_transformed = mds_nonmetric.fit_transform(data_sample)

# Wizualizacja niemetrycznego MDS
plt.figure(figsize=(8, 6))
plt.scatter(mds_nonmetric_transformed[:, 0], mds_nonmetric_transformed[:, 1], alpha=0.6)
plt.title("Niemetryczne skalowanie wielowymiarowe (MDS)")
plt.xlabel("Wymiar 1 (Dim1)")
plt.ylabel("Wymiar 2 (Dim2)")
plt.grid(True)
plt.show()
```

```python
# Niemetryczne MDS z 3 wymiarami
mds_nonmetric = MDS(n_components=3, random_state=42, metric=False)
mds_nonmetric_transformed = mds_nonmetric.fit_transform(data_sample)

# Tworzenie interaktywnego wykresu 3D
fig = go.Figure()

# Dodanie punktów do wykresu
fig.add_trace(
    go.Scatter3d(
        x=mds_nonmetric_transformed[:, 0],  # Wymiar 1
        y=mds_nonmetric_transformed[:, 1],  # Wymiar 2
        z=mds_nonmetric_transformed[:, 2],  # Wymiar 3
        mode='markers',
        marker=dict(
            size=1,
            color=np.arange(len(mds_nonmetric_transformed)),  # Kolory w zależności od indeksu
            colorscale='Plasma',  # Skala kolorów
            opacity=0.5
        ),
        text=[f"Punkt {i+1}" for i in range(len(mds_nonmetric_transformed))],  # Etykiety punktów
        name='Punkty'
    )
)

# Ustawienia osi
fig.update_layout(
    scene=dict(
        xaxis_title="Wymiar 1 (Dim1)",
        yaxis_title="Wymiar 2 (Dim2)",
        zaxis_title="Wymiar 3 (Dim3)",
    ),
    title="Interaktywny Wykres 3D - Niemetryczne MDS",
    margin=dict(l=0, r=0, b=0, t=40),
)

# Wyświetlenie wykresu
fig.show()
```

Opis metod
1. PCA (Principal Component Analysis)
Opis: PCA to metoda liniowej redukcji wymiarowości, która znajduje nowe osie (składowe główne) maksymalizujące wariancję danych. Wykorzystuje dekompozycję macierzy kowariancji lub korelacji.
Zalety:
Zachowuje jak najwięcej informacji (wariancji).
Prosta i szybka implementacja.
Wady:
Zakłada liniowość zależności.
Trudna interpretacja nowych składowych jako kombinacji liniowych.
2. Klasyczne MDS
Opis: Klasyczne MDS odwzorowuje rzeczywiste odległości w przestrzeni niskowymiarowej, minimalizując różnice między oryginalnymi odległościami a odwzorowanymi.
Zalety:
Umożliwia intuicyjną wizualizację relacji przestrzennych.
Zachowuje metryczne zależności.
Wady:
Wymaga danych w formie macierzy odległości.
Wrażliwa na szum w danych.
3. Niemetryczne MDS
Opis: Niemetryczne MDS odwzorowuje porządek rangowy odległości między obiektami, ignorując ich dokładne wartości.
Zalety:
Przydatna dla danych nieliniowych i skal porządkowych.
Elastyczniejsza niż klasyczne MDS.
Wady:
Wyższy koszt obliczeniowy.
Wyniki zależą od początkowych warunków iteracyjnych.


```python
plt.figure(figsize=(24, 6))

# PCA
plt.subplot(1, 3, 1)
plt.scatter(pca_transformed[:, 0], pca_transformed[:, 1], alpha=0.6)
plt.title("PCA (PC1 vs PC2)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(True)

# Klasyczne MDS
plt.subplot(1, 3, 2)
plt.scatter(mds_transformed[:, 0], mds_transformed[:, 1], alpha=0.6)
plt.title("Klasyczne MDS (Dim1 vs Dim2)")
plt.xlabel("Dim1")
plt.ylabel("Dim2")
plt.grid(True)

# Niemetryczne MDS
plt.subplot(1, 3, 3)
plt.scatter(mds_nonmetric_transformed[:, 0], mds_nonmetric_transformed[:, 1], alpha=0.6)
plt.title("Niemetryczne MDS (Dim1 vs Dim2)")
plt.xlabel("Dim1")
plt.ylabel("Dim2")
plt.grid(True)

plt.tight_layout()
plt.show()
```

Wyniki porównania PCA, klasycznego MDS i niemetrycznego MDS:
1. PCA (Principal Component Analysis)
Dane w przestrzeni dwóch pierwszych składowych głównych (PC1 i PC2) wykazują wyraźne grupowanie obserwacji.
PCA dobrze zachowuje zmienność danych i ujawnia dominujące kierunki wariancji.
Obserwacje leżące daleko od środka mogą wskazywać na punkty odstające lub specyficzne wzorce w danych.
2. Klasyczne MDS
Dane zostały odwzorowane na podstawie rzeczywistych odległości euklidesowych.
Układ punktów jest podobny do PCA, ale bardziej równomiernie rozłożony, co może być korzystne w analizach relacji przestrzennych.
3. Niemetryczne MDS
Układ punktów w niemetrycznym MDS różni się od klasycznego, zachowując jedynie porządek rangowy odległości.
Metoda lepiej odwzorowuje nieliniowe zależności w danych, co może być przydatne w przypadku nieregularnych struktur.
Wnioski:
PCA jest najskuteczniejsze w zachowaniu globalnej wariancji danych i ujawnianiu głównych kierunków zmienności.
Klasyczne MDS najlepiej odwzorowuje rzeczywiste odległości między punktami, co jest przydatne przy analizie przestrzennej.


Uzasadnienie wyboru zmiennych
W projekcie uwzględniono następujące zmienne:

Study_Hours_Per_Day: Czas poświęcony na naukę.
Extracurricular_Hours_Per_Day: Czas poświęcony na zajęcia dodatkowe.

Stress_Level: Poziom stresu (zmienna docelowa).
Dlaczego wybraliśmy te zmienne?
1. Znaczenie zmiennych dla analizy
Poziom stresu (Stress_Level) jest zmienną docelową, kluczową dla naszego projektu.
Pozostałe zmienne opisują różne aspekty stylu życia studentów (akademicki, społeczny, fizyczny), które mogą wpływać na poziom stresu.
2. Korelacje między zmiennymi

Ogólne relacje:
Styl życia studentów, który równoważy czas przeznaczony na naukę, sen, aktywność fizyczną i życie społeczne, może pozytywnie wpływać na poziom stresu.
Studenci, którzy mają lepszą organizację czasu (więcej nauki i wyższe GPA), wydają się być mniej zestresowani.
Wnioski na temat poziomu stresu studentów
Główne czynniki wpływające na poziom stresu to czas poświęcany na naukę i osiągnięcia akademickie.
Sen, aktywność fizyczna i życie społeczne odgrywają mniejszą, ale nadal istotną rolę w zarządzaniu stresem.
Wysoki poziom stresu może być związany z brakiem równowagi w stylu życia, np. zbyt mało czasu na naukę lub brak aktywności fizycznej.
